# Load datasets

# Average data according to month (x12 samples per site)

```python
# Extract month for grouping, drop the date column, and calculate monthly mean for each dataframe
BF1['month'] = BF1['date'].dt.month
BF1 = BF1.drop(columns=['date'])
monthly_means_BF1 = BF1.groupby('month').mean()
monthly_means_BF1.to_csv('monthly_means_BF1.csv')

BF2['month'] = BF2['date'].dt.month
BF2 = BF2.drop(columns=['date'])
monthly_means_BF2 = BF2.groupby('month').mean()
monthly_means_BF2.to_csv('monthly_means_BF2.csv')

CP3['month'] = CP3['date'].dt.month
CP3 = CP3.drop(columns=['date'])
monthly_means_CP3 = CP3.groupby('month').mean()
monthly_means_CP3.to_csv('monthly_means_CP3.csv')

CP6['month'] = CP6['date'].dt.month
CP6 = CP6.drop(columns=['date'])
monthly_means_CP6 = CP6.groupby('month').mean()
monthly_means_CP6.to_csv('monthly_means_CP6.csv')

CP9['month'] = CP9['date'].dt.month
CP9 = CP9.drop(columns=['date'])
monthly_means_CP9 = CP9.groupby('month').mean()
monthly_means_CP9.to_csv('monthly_means_CP9.csv')

EW1['month'] = EW1['date'].dt.month
EW1 = EW1.drop(columns=['date'])
monthly_means_EW1 = EW1.groupby('month').mean()
monthly_means_EW1.to_csv('monthly_means_EW1.csv')

EW2['month'] = EW2['date'].dt.month
EW2 = EW2.drop(columns=['date'])
monthly_means_EW2 = EW2.groupby('month').mean()
monthly_means_EW2.to_csv('monthly_means_EW2.csv')

EZU['month'] = EZU['date'].dt.month
EZU = EZU.drop(columns=['date'])
monthly_means_EZU = EZU.groupby('month').mean()
monthly_means_EZU.to_csv('monthly_means_EZU.csv')

JHK['month'] = JHK['date'].dt.month
JHK = JHK.drop(columns=['date'])
monthly_means_JHK = JHK.groupby('month').mean()
monthly_means_JHK.to_csv('monthly_means_JHK.csv')

MB1['month'] = MB1['date'].dt.month
MB1 = MB1.drop(columns=['date'])
monthly_means_MB1 = MB1.groupby('month').mean()
monthly_means_MB1.to_csv('monthly_means_MB1.csv')

MB2['month'] = MB2['date'].dt.month
MB2 = MB2.drop(columns=['date'])
monthly_means_MB2 = MB2.groupby('month').mean()
monthly_means_MB2.to_csv('monthly_means_MB2.csv')

MCP['month'] = MCP['date'].dt.month
MCP = MCP.drop(columns=['date'])
monthly_means_MCP = MCP.groupby('month').mean()
monthly_means_MCP.to_csv('monthly_means_MCP.csv')

MLP['month'] = MLP['date'].dt.month
MLP = MLP.drop(columns=['date'])
monthly_means_MLP = MLP.groupby('month').mean()
monthly_means_MLP.to_csv('monthly_means_MLP.csv')

SKU['month'] = SKU['date'].dt.month
SKU = SKU.drop(columns=['date'])
monthly_means_SKU = SKU.groupby('month').mean()
monthly_means_SKU.to_csv('monthly_means_SKU.csv')
```
