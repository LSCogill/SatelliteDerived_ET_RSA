# All subplot

```python
# Import necessary libraries
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Load the data for each dataset (assuming you've already each sites data)
BF1 = pd.read_csv('ET_Monthly_BF1.csv')
BF2 = pd.read_csv('ET_Monthly_BF2.csv')
CP3 = pd.read_csv('ET_Monthly_CP3.csv')
CP6 = pd.read_csv('ET_Monthly_CP6.csv')
CP9 = pd.read_csv('ET_Monthly_CP9.csv')
EW1 = pd.read_csv('ET_Monthly_EW1.csv')
EW2 = pd.read_csv('ET_Monthly_EW2.csv')
EZU = pd.read_csv('ET_Monthly_EZU.csv')
JHK = pd.read_csv('ET_Monthly_JHK.csv')
MB1 = pd.read_csv('ET_Monthly_MB1.csv')
MB2 = pd.read_csv('ET_Monthly_MB2.csv')
MCP = pd.read_csv('ET_Monthly_MCP.csv')
MLP = pd.read_csv('ET_Monthly_MLP.csv')
SKU = pd.read_csv('ET_Monthly_SKU.csv')
```

```python
# Convert 'Month_Year' to datetime for each dataset
BF1['Month_Year'] = pd.to_datetime(BF1['Month_Year'], format="%d/%m/%Y")
BF2['Month_Year'] = pd.to_datetime(BF2['Month_Year'], format="%d/%m/%Y")
CP3['Month_Year'] = pd.to_datetime(CP3['Month_Year'], format="%d/%m/%Y")
CP6['Month_Year'] = pd.to_datetime(CP6['Month_Year'], format="%d/%m/%Y")
CP9['Month_Year'] = pd.to_datetime(CP9['Month_Year'], format="%d/%m/%Y")
EW1['Month_Year'] = pd.to_datetime(EW1['Month_Year'], format="%d/%m/%Y")
EW2['Month_Year'] = pd.to_datetime(EW2['Month_Year'], format="%d/%m/%Y")
EZU['Month_Year'] = pd.to_datetime(EZU['Month_Year'], format="%d/%m/%Y")
JHK['Month_Year'] = pd.to_datetime(JHK['Month_Year'], format="%d/%m/%Y")
MB1['Month_Year'] = pd.to_datetime(MB1['Month_Year'], format="%d/%m/%Y")
MB2['Month_Year'] = pd.to_datetime(MB2['Month_Year'], format="%d/%m/%Y")
MCP['Month_Year'] = pd.to_datetime(MCP['Month_Year'], format="%d/%m/%Y")
MLP['Month_Year'] = pd.to_datetime(MLP['Month_Year'], format="%d/%m/%Y")
SKU['Month_Year'] = pd.to_datetime(SKU['Month_Year'], format="%d/%m/%Y")
```

```python
# Define custom titles for each subplot
subplot_titles = [
    '(a) Benfontein - Site 1', '(b) Cathedral Peak - Catchment 3',
    '(c) Cathedral Peak - Catchment 6', '(d) Cathedral Peak - Catchment 9', '(e) Endwell - Site 1',
    '(f) Endwell - Site 2', '(g) Ezulu', '(h) Middelburg - Site 1',
    '(i) Middelburg - Site 2', '(l) Maputaland Coastal Plain', '(m) Malopeni',
    '(n) Skukuza'
]

# Create the subplot grid with 5 rows and 3 columns
fig = make_subplots(
    rows=5, cols=3,
    subplot_titles=subplot_titles,  # Use custom titles for each subplot
    shared_xaxes=False, shared_yaxes=True,  # Share axes across plots
    vertical_spacing=0.05, horizontal_spacing=0.05
)
```

```python
# Add traces for each dataset
datasets = [BF1, BF2, CP3, CP6, CP9, EW1, EW2, EZU, JHK, MB1, MB2, MCP, MLP, SKU]  # Add all datasets (BF1 to SKU)
for i, dataset in enumerate(datasets):
    row = i // 3 + 1  # Row index for subplot (1-based)
    col = i % 3 + 1   # Column index for subplot (1-based)

    # Add traces for different datasets in each subplot
    fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['Field_ET_monthly'], mode='lines', name='Field_ET', line=dict(color='black', width=4)), row=row, col=col)
    fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['TC_monthly'], mode='lines', name='TerraClimate_ET', line=dict(color='blue')), row=row, col=col)
    fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['SMAP_monthly'], mode='lines', name='SMAP_ET', line=dict(color='orange')), row=row, col=col)
    fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['GLDAS_monthly'], mode='lines', name='GLDAS_ET', line=dict(color='red')), row=row, col=col)
    fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['FLDAS_monthly'], mode='lines', name='FLDAS_ET', line=dict(color='purple')), row=row, col=col)
    fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['WaPOR_monthly'], mode='lines', name='WaPOR_ET', line=dict(color='limegreen')), row=row, col=col)
    fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['MOD16_monthly'], mode='lines', name='MOD16_ET', line=dict(color='saddlebrown')), row=row, col=col)

    if 'SSEBop_monthly' in dataset.columns:
        fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['SSEBop_monthly'], mode='lines', name='SSEBop_ET', line=dict(color='magenta')), row=row, col=col)
    if 'PTjpl_monthly' in dataset.columns:
        fig.add_trace(go.Scatter(x=dataset['Month_Year'], y=dataset['PTjpl_monthly'], mode='lines', name='PTjpl_ET', line=dict(color='yellow')), row=row, col=col)

    # Customize the x-axis date range for each subplot (optional: set specific date range for each)
    fig.update_xaxes(
        range=[min(dataset['Month_Year']), max(dataset['Month_Year'])],  # Set the range based on the dataset's date range
        row=row, col=col,
        dtick="M1",  # Set tick every month
        tickformat="%b %Y",  # Format to show Month-Year
        tickfont=dict(size=6),  # Set font size to 10
        tickangle=-45,  # Set angle to 45 degrees
        showgrid=False,
        showline=True,  # Show line at the bottom
        linecolor='black',  # Border color
    )
 # Set the y-axis limit for each subplot to 150
    fig.update_yaxes(
        range=[0, 150],  # Set the y-axis range to [0, 150]
        row=row, col=col,
        showgrid=False,
        showline=True,  # Show line at the left
        linecolor='black',)

# Adjusting the layout and aspect ratio for the subplots
fig.update_layout(
    title='',
    yaxis=dict(
        title='',
        tickfont=dict(size=11),
        showgrid=False,
        title_standoff=10
    ),
    xaxis=dict(
        title='',  # Title for the x-axis
        tickfont=dict(size=11),
        showgrid=False,
        title_standoff=10
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    showlegend=False,
    height=800,  # Adjust the height as necessary for more rectangular plots
    width=1000,  # Adjust the width to make the overall plot more rectangular

    title_text="",  # Global plot title
)

# Show the rough plot
fig.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Assuming your datasets are already loaded and the 'Month_Year' column is in the 'Month-Year' format.

# Titles for each subplot
titles = [
    '(a) Benfontein - Site 1', '(b) Benfontein - Site 2', '(c) Cathedral Peak - Catchment 3',
    '(d) Cathedral Peak - Catchment 6', '(e) Cathedral Peak - Catchment 9', '(f) Endwell - Site 1',
    '(g) Endwell - Site 2', '(h) Ezulu', '(i) Jonkershoek', '(j) Middelburg - Site 1',
    '(k) Middelburg - Site 2', '(l) Maputaland Coastal Plain', '(m) Malopeni', '(n) Skukuza'
]

# Create a 5x3 subplot structure
fig, axes = plt.subplots(5, 3, figsize=(17, 10))
axes = axes.flatten()  # Flatten to easily iterate over all subplots

# Define the color scheme for the variables
colors = {
    'Field_ET_monthly': 'black',
    'TC_monthly': 'blue',
    'SMAP_monthly': 'orange',
    'SSEBop_monthly': 'magenta',
    'GLDAS_monthly': 'red',
    'FLDAS_monthly': 'purple',
    'WaPOR_monthly': 'limegreen',
    'MOD16_monthly': 'saddlebrown',
    'PTjpl_monthly': 'yellow'
}
for i, df in enumerate(datasets):
    # Ensure 'Month_Year' is in datetime format
    df['Month_Year'] = pd.to_datetime(df['Month_Year'], format='%b %Y')

    ax = axes[i]

    # Plot each dataset with a specific color
    for col, color in colors.items():
        if col in df.columns:  # Ensure the column exists in the dataset
            ax.plot(df['Month_Year'], df[col], linestyle='-', color=color, label=col)

    # Force the Field_ET line to plot last, so it isn't overwritten
    if 'Field_ET' in df.columns:
        ax.plot(df['Month_Year'], df['Field_ET'], color='black', linestyle='-', label='Field_ET', zorder=10)

    # Rest of your loop code...

for i, df in enumerate(datasets):
    # Ensure 'Month_Year' is in datetime format
    df['Month_Year'] = pd.to_datetime(df['Month_Year'], format='%b %Y')

    ax = axes[i]

    # Plot each dataset with a specific color
    for col, color in colors.items():
        if col in df.columns:  # Ensure the column exists in the dataset
            ax.plot(df['Month_Year'], df[col], linestyle='-', color=color, label=col)

    # Set custom title for each subplot
    ax.set_title(titles[i], fontsize=12)
    ax.set_xlabel('Month-Year', fontsize=10)
    ax.set_ylabel('Evapotranspiration (mm/month)', fontsize=10)

    # Rotate x-axis labels to -45 degrees
    ax.tick_params(axis='x', labelsize=6, rotation=45)
    for label in ax.get_xticklabels():
        label.set_rotation(45)
        label.set_ha('right')  # Align labels to the left for better readability

        # Set specific y-axis limit for the 12th subplot
    if i == 11:
        ax.set_ylim(0, 150)  # Set y-axis limit to 120 for the 11th subplot


    # Set x-axis interval for subplots
    if i in [2, 3, 4, 8, 9, 10]:  # 1st, 3rd, 5th, and 9th subplots (indexed as 0, 2, 4, 8)
        ax.set_xticks(df['Month_Year'][::3])  # Select every 3rd date
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Display Month-Year format
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # Set 3-month interval
    elif i in [0, 5, 6, 12, 13]:  # 5th and 6th subplots (indexed as 4 and 5)
        ax.set_xticks(df['Month_Year'][::2])  # Select every 2nd date
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Display Month-Year format
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))  # Set 2-month interval
    else:  # For all other subplots (2nd, 4th, 6th, etc.)
        ax.set_xticks(df['Month_Year'][::1])  # Select every 2nd date
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Display Month-Year format
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))  # Set 1-month interval

    # Remove individual subplot x and y labels (optional)
    ax.set_xlabel('')
    ax.set_ylabel('')

# Set overarching x-axis and y-axis labels
fig.supxlabel('Date', fontsize=14)
fig.supylabel('Evapotranspiration (mm/month)', fontsize=14, x=0.01)

# Hide any unused subplots
for j in range(len(datasets), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
